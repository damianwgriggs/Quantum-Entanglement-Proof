```python
# --- PROTOCOL Q-1: FULL ENTANGLEMENT ARCHITECT SCRIPT (CORRECTED) ---

# 1. Imports
import cirq
import numpy as np
import matplotlib.pyplot as plt

# 2. Define Qubits
# The variable names are used directly by the gates now.
qubit_A = cirq.LineQubit(0)
qubit_B = cirq.LineQubit(1)

# 3. Create the Entangled Circuit (Bell State)
circuit = cirq.Circuit()

# Use the cleaner gate application syntax: Gate(*qubits)
# H Gate on A (Superposition)
circuit.append(cirq.H(qubit_A))

# CNOT Gate controlled by A, targeting B (Entanglement)
# The key fix: CNOT(control_qubit, target_qubit)
circuit.append(cirq.CNOT(qubit_A, qubit_B))

# Final Measurement
circuit.append(cirq.measure(qubit_A, qubit_B, key='result'))


# Output the Circuit Diagram (Success Criterion 1)
print("--- Circuit Diagram (ASCII) ---")
print(circuit)
print("\n" + "="*50 + "\n")


# 4. Execute the Circuit on the Quantum Virtual Machine (QVM)
repetitions = 1000
simulator = cirq.Simulator()
results = simulator.run(circuit, repetitions=repetitions)


# 5. Process and Visualize Results (Success Criterion 2)
# Get the histogram of outcomes (counts)
counts = results.histogram(key='result')

print(f"--- Quantum Measurement Results (Counts out of {repetitions}) ---")
print(counts)
print("\n" + "="*50)

# Generate the Histogram Plot
cirq.plot_state_histogram(counts, title=f"Entanglement Verification (Bell State) after {repetitions} Runs")
plt.show()

print("\nPROJECT Q-1 SUCCESS: Entanglement Proof Achieved.")
```
